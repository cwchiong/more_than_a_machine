```python
import numpy
import matplotlib
import pylab

x = numpy.linspace(0.000000001, 1, 1000)
y = numpy.tanh(-5 * numpy.log(x) * numpy.log(x) * numpy.log(x))
# The logarithm allows high values inputs (creativity, reasoning, and manipulation) will output values close to zero.
# The hyperbolic tangent insures that the results of our model stays between zero and one as it is a percentage.
pylab.plot(x, y)
pylab.show()
```

```python
from statistics import mean
import numpy
import matplotlib

xs = numpy.array([0.0853, 0.0477, 0, 0.8966, 0.9987, 0.0370, 0.9766, 0.0225, 0.0853, 0.1514], dtype=numpy.float64)
ys = numpy.array([0.23, 0.019, 0.012, 0.94, 0.96, 0.0074, 0.97, 0.009, 0.03, 0.17], dtype=numpy.float64)
# xs being the values calculated using our model, and ys being the values calculated from the Frey & Osborne model

def slope_intercept(xs, ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    b = mean(ys) - m*mean(xs)
    
    return m, b

m, b = slope_intercept(xs,ys)
regression_line = [(m*x)+b for x in xs]
pylab.scatter(xs, ys)
pylab.plot(xs, regression_line)
pylab.show()
```
